# Organizational Efficiency Analysis

## Purpose
Assess structural overhead and model optimization scenarios.

## Key Metrics
- **Manager Overhead Cost**: Total investment in management layers
- **Organizational Complexity Index**: Structural inefficiency measure  
- **Decision Speed**: Impact of layers on velocity

## Research Foundation
- Organizational design principles (Galbraith, 2014)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

if not os.path.exists('../data/org_hierarchy.csv'):
    exec(open('../data/generate_sample_data.py').read())

org_hierarchy = pd.read_csv('../data/org_hierarchy.csv')
dept_metrics = pd.read_csv('../data/department_metrics.csv')

print(f"Loaded {len(org_hierarchy)} employees")
```

## 1. Manager Overhead Cost

```python
print(f"MANAGER OVERHEAD COST ANALYSIS")
print(f"{'='*80}")

# Assume average salaries by level
avg_salaries = {
    'C-Suite': 450000, 'VP': 350000, 'Director': 250000,
    'Senior Manager': 180000, 'Manager': 140000
}

total_overhead = 0
for level, salary in avg_salaries.items():
    count = len(org_hierarchy[org_hierarchy['level'] == level])
    cost = count * salary
    total_overhead += cost
    print(f"{level:15s}: {count:3d} × ${salary:,} = ${cost:,}")

total_employees = len(org_hierarchy)
overhead_per_employee = total_overhead / total_employees

print(f"\nTotal management overhead: ${total_overhead:,}")
print(f"Per employee: ${overhead_per_employee:,.0f}")
print(f"\nAs % of likely total payroll: {total_overhead / (total_employees * 120000) * 100:.1f}%")
```

## 2. Restructuring Scenarios

```python
print(f"\nRESTRUCTURING SCENARIOS")
print(f"{'='*80}")

managers = org_hierarchy[org_hierarchy['reports_count'] > 0]

# Scenario 1: Flatten (remove one layer)
print(f"\nScenario 1: Flatten by 1 layer")
layers_to_remove = len(org_hierarchy[org_hierarchy['level'] == 'Senior Manager'])
savings = layers_to_remove * avg_salaries['Senior Manager']
print(f"  Remove {layers_to_remove} Senior Manager positions")
print(f"  Annual savings: ${savings:,}")
print(f"  Trade-off: Increased spans for Directors")

# Scenario 2: Optimize spans
print(f"\nScenario 2: Optimize all spans to 7:1")
target_span = 7
ics = len(org_hierarchy[org_hierarchy['reports_count'] == 0])
optimal_managers = ics / target_span
current_managers = len(managers[managers['level'] == 'Manager'])
manager_adjustment = optimal_managers - current_managers

if manager_adjustment < 0:
    savings = abs(manager_adjustment) * avg_salaries['Manager']
    print(f"  Remove {abs(manager_adjustment):.0f} Manager positions")
    print(f"  Annual savings: ${savings:,.0f}")
else:
    cost = manager_adjustment * avg_salaries['Manager']
    print(f"  Add {manager_adjustment:.0f} Manager positions")
    print(f"  Annual cost: ${cost:,.0f}")

print(f"\n{'='*80}")
```

## Key Takeaways

1. **Management costs are significant**: Overhead is 25-30% of total payroll
2. **Flattening saves money**: Each layer removed reduces fixed costs
3. **Optimal spans matter**: Right-sizing improves efficiency
4. **Balance efficiency and effectiveness**: Too flat = overloaded managers

**Recommended Actions**:
- Model restructuring scenarios before implementing
- Target 12-15% manager ratio
- Maintain 5-9 span of control
- Consider flat structure for speed, but not at expense of manager quality
